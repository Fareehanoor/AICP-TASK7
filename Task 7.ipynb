```python
# TASK 1 - Set up the donation system


charities = ["Charity A", "Charity B", "Charity C"]
charity_totals = [0, 0, 0]


print("Choose a charity:")
for i, charity in enumerate(charities, start=1):
    print(f"{i}. {charity}")


charity_choice = int(input("Enter the number of your chosen charity (1, 2, or 3): "))

if charity_choice not in [1, 2, 3]:
    print("Invalid choice. Please choose 1, 2, or 3.")
    exit()

shopping_bill = float(input("Enter the shopping bill amount: "))


donation = shopping_bill * 0.01


charity_totals[charity_choice - 1] += donation


# TASK 2 - Record and total each donation

def record_and_total_donation():
    # Get user input for charity choice or -1 to show totals
    charity_choice = int(input("Enter the number of your chosen charity (1, 2, 3), or -1 to show totals: "))

    # Check if user wants to show totals
    if charity_choice == -1:
        show_totals()
        return


    if charity_choice not in [1, 2, 3]:
        print("Invalid choice. Please choose 1, 2, or 3.")
        return


    shopping_bill = float(input("Enter the shopping bill amount: "))


    donation = shopping_bill * 0.01


    charity_totals[charity_choice - 1] += donation


    print(f"Donation of {donation:.2f} recorded for {charities[charity_choice - 1]}.")


# TASK 3 - Show the totals so far

def show_totals():

    combined_data = list(zip(charities, charity_totals))


    combined_data.sort(key=lambda x: x[1], reverse=True)


    print("\nCharity Totals:")
    for charity, total in combined_data:
        print(f"{charity}: {total:.2f}")

    # Calculate and display the grand total
    grand_total = sum(charity_totals)
    print(f"\nGRAND TOTAL DONATED TO CHARITY: {grand_total:.2f}")


# Test the program
record_and_total_donation()
record_and_total_donation()
record_and_total_donation()
record_and_total_donation()
record_and_total_donation()
```
